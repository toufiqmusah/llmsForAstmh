```python
grouped_df = df.groupby(['generalCategory'])['mergedCategory'].unique()
```

```python
mergedtoGen = dict(grouped_df)
```

```python
mergedIdxs = pd.DataFrame(df['mergedCategory'].unique(), columns=['names'])
```

```python
mergedIdxs
```

```python
total_corr = 0

for i, row in df.iterrows():
    predidx = row['prediction']
    pred_cat = mergedIdxs.iloc[predidx]
    trueidx = row['label']
    true_cat = mergedIdxs.iloc[trueidx]

    k_true = {j for j in mergedtoGen if true_cat.values[0] in mergedtoGen[j]}
    k_pred = {j for j in mergedtoGen if pred_cat.values[0] in mergedtoGen[j]}

    if k_pred == k_true:
        total_corr += 1
```
